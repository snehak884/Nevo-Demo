```python
import pandas as pd
```

```python
df = pd.read_csv("../../collector_test_results.csv")
df["true_pos_rate"] = df["correct_count"] / df["total_count"]
df["false_pos_rate"] = df["false_pos_count"] / df["total_count"]
df["accuracy"] = (df["correct_count"] - df["false_pos_count"]) / df["total_count"]
df.head()
```

Look into the falso positive samples.

```python
for _, row in df.loc[df["false_pos_count"]>0].iterrows():
    print(row["user_message"])
    print(".\n".join(str(t) for t in eval(row["agent_data"])))
    print()
```

```python
print("accuracy by model")
display(df.groupby(["agent"])["accuracy"].mean().sort_values(ascending=False))

print("true_pos_rate by model")
display(df.groupby(["agent"])["true_pos_rate"].mean().sort_values(ascending=False))

print("false_pos_rate pos by model")
display(df.groupby(["agent"])["false_pos_rate"].mean().sort_values(ascending=False))

print("Accuracy by model and repetition – there are variations between repetitions")
display(pd.pivot_table(df, index="agent", values="accuracy", columns="repetition", aggfunc="mean").sort_index())
```
