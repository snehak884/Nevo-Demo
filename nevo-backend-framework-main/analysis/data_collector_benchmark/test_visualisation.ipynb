```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
four_o_mini_single = pd.read_csv('test-4o-mini-single20.csv')
four_o_mini_multiple = pd.read_csv('test-4o-mini-multi20.csv')
four_o_single = pd.read_csv('test-4o-single20.csv')

def plot_times(df1: pd.DataFrame, df2: pd.DataFrame, name1: str, name2: str):    
    fig, ax = plt.subplots(1,3, figsize=(15,5))

    ax[0].hist(df1['mean_duration_s'], alpha=0.5, label=f'{name1} Collectors', density=True)
    ax[0].hist(df2['mean_duration_s'], alpha=0.5, label=f'{name2} Collectors', density=True)
    ax[0].set_xlabel('Mean Duration (s)')
    ax[0].set_ylabel('Frequency')
    ax[0].set_title(f"Mean duration by run of {name1} collectors vs {name2} collectors", size=10)
    ax[0].legend()

    ax[1].hist(df1['max_duration_s'], alpha=0.5, label=f'{name1} Collectors', density=True)
    ax[1].hist(df2['max_duration_s'], alpha=0.5, label=f'{name2} Collectors', density=True)
    ax[1].set_xlabel('Max Duration (s)')
    ax[1].set_ylabel('Frequency')
    ax[1].set_title(f"Max duration by run of {name1} collectors vs {name2} collectors", size=10)
    ax[1].legend()


    ax[2].hist(df1['min_duration_s'], alpha=0.5, label=f'{name1} Collectors', density=True)
    ax[2].hist(df2['min_duration_s'], alpha=0.5, label=f'{name2} Collectors', density=True)
    ax[2].set_xlabel('Min Duration (s)')
    ax[2].set_ylabel('Frequency')
    ax[2].set_title(f"Min duration by run of {name1} collectors vs {name2} collectors", size=10)
    ax[2].legend()

    plt.tight_layout()
    plt.show()

def plot_metrics(df1: pd.DataFrame, df2: pd.DataFrame, name1: str, name2: str):    
    fig, ax = plt.subplots(1,3, figsize=(15,5))

    ax[0].hist(df1['precision'], alpha=0.5, label=f'{name1} Collectors', density=True)
    ax[0].hist(df2['precision'], alpha=0.5, label=f'{name2} Collectors', density=True)
    ax[0].set_xlabel('Precision')
    ax[0].set_ylabel('Frequency')
    ax[0].set_title(f"Mean precision by run of {name1} collectors vs {name2} collectors", size=10)
    ax[0].legend()

    ax[1].hist(df1['recall'], alpha=0.5, label=f'{name1} Collectors', density=True)
    ax[1].hist(df2['recall'], alpha=0.5, label=f'{name2} Collectors', density=True)
    ax[1].set_xlabel('Recall')
    ax[1].set_ylabel('Frequency')
    ax[1].set_title(f"Mean recall by run of {name1} collectors vs {name2} collectors", size=10)
    ax[1].legend()


    ax[2].hist(df1['f1'], alpha=0.5, label=f'{name1} Collectors', density=True)
    ax[2].hist(df2['f1'], alpha=0.5, label=f'{name2} Collectors', density=True)
    ax[2].set_xlabel('F1')
    ax[2].set_ylabel('Frequency')
    ax[2].set_title(f"Mean F1 by run of {name1} collectors vs {name2} collectors", size=10)
    ax[2].legend()

    plt.tight_layout()
    plt.show()

plot_times(four_o_mini_single, four_o_single, name1='4o-mini', name2='4o')
plot_metrics(four_o_mini_single, four_o_single, name1='4o-mini', name2='4o')

plot_times(four_o_mini_single, four_o_mini_multiple, name1='4o-mini-single', name2='4o-mini-multi')
plot_metrics(four_o_mini_single, four_o_mini_multiple, name1='4o-mini-single', name2='4o-mini-multi')
```
